```python
def bar_plot(df, kategorik_vars, target_var):
    cat           =  df[kategorik_vars]
    n             =  len(cat.columns)
    rows          =  n // 2 + n % 2
    cols          =  2
    sns.set(style = "whitegrid", font_scale=1)
    fig, ax       =  plt.subplots(rows, cols, figsize=(20, rows * 5))

    for i in range(rows):
        for j in range(cols):
            index    = i * cols + j
            
            if index < n:
                col  = cat.columns[index]
                palette = sns.color_palette(colors)
                
                sns.countplot(ax=ax[i, j],
                              data=df,
                              x=col,
                              color=colors,
                              hue=target_var,
                              palette=palette)
                
                ax[i, j].set_title(f'Hubungan antara {col} terhadap {target_var}', fontdict={'fontsize': 15})
                ax[i, j].set_xlabel(col)
                ax[i, j].set_ylabel('Count')

    plt.tight_layout()
    plt.show()
```

```python
def num_hist_plot(df, numerical_columns, target_var):
    num           =  df[numerical_columns]
    n             =  len(num.columns)
    rows          =  n // 2 + n % 2
    cols          =  2
    sns.set(style = "whitegrid", font_scale=1)
    fig, ax       =  plt.subplots(rows, cols, figsize=(20, rows * 5))

    for i in range(rows):
        for j in range(cols):
            index    = i * cols + j
            
            if index < n:
                col  = num.columns[index]
                palette = sns.color_palette(colors)  # define your palette here
                
                sns.histplot(ax=ax[i, j],
                              data=df,
                              x=col,
                              hue=target_var,
                              palette=palette)
                
                ax[i, j].set_title(f'Distribusi {col} terhadap {target_var}', fontdict={'fontsize': 15})
                ax[i, j].set_xlabel(col)
                ax[i, j].set_ylabel('Count')

    plt.tight_layout()
    plt.show()
```

1. **Extended Isolation Forest** adalah peningkatan dari Isolation Forest yang memperbaiki cara pohon-pohon isolasi dibangun dan dipotong, sehingga meningkatkan kinerja dalam mendeteksi anomali. EIF menggunakan hyperplane (bukan hanya garis) untuk membagi data pada setiap node, yang memungkinkan deteksi anomali yang lebih presisi.

2. **Local Outlier Factor (LOF)** adalah metode deteksi outlier yang unsupervised dengan menghitung deviasi kepadatan lokal dari suatu titik data dengan tetangganya. LOF membandingkan kepadatan lokal dari suatu titik data dengan kepadatan tetangganya. Jika suatu titik data memiliki kepadatan yang jauh lebih rendah dibandingkan dengan tetangganya, maka titik data tersebut dianggap sebagai outlier.

3. **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** adalah algoritma klasterisasi berbasis kepadatan yang tidak memerlukan parameter khusus jumlah klaster. Algoritma ini mengelompokkan titik-titik yang berdekatan dan menandai titik-titik yang berada di wilayah dengan kepadatan rendah sebagai outlier.

```python
print("Klasifikasi Anomali Report EIF Terhadap Data Test")
print(classification_report(y_test, y_pred_test_ifo))
```

Adjusted Rand Index (ARI)

Adjusted Rand Index (ARI) adalah sebuah metrik yang digunakan untuk mengevaluasi kesamaan antara dua klaster atau pengelompokan data. Ini digunakan untuk mengukur kualitas hasil clustering atau pengelompokan data dengan membandingkan dengan label ground truth (label asli).

Adjusted Rand Index (ARI) mengukur kesamaan antara dua klaster (pengelompokan data). Nilai ARI berkisar antara -1 hingga 1:
- Nilai 1: Hasil clustering identik dengan label asli.
- Nilai 0: Hasil clustering tidak lebih baik daripada pengelompokan acak.
- Nilai negatif: Hasil clustering lebih buruk daripada pengelompokan acak.

```python
# Adjusted Rand Index

adjusted_rand_score(y_train, y_pred_train_ifo), adjusted_rand_score(y_test, y_pred_test_ifo)
```

Hasil Klasifikasi Report dan Rand Score menunjukkan performa EIF yang cukup bagus dan tidak menunjukkan adanya overvit? Artinya model mungkin cocok dalam mendeteksi outlier yang merupakan minoritas.
